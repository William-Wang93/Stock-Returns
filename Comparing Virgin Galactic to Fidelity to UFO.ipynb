```python
import yfinance as yf
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from datetime import datetime

```

```python
UFO = yf.Ticker('UFO')
UFO_history=pd.DataFrame(UFO.history(period='max'))
UFO_history_close = pd.DataFrame(UFO_history['Close'])
#df= UFO_history_close.iloc[:,0:1]
UFO_Close_dates = pd.DataFrame(UFO_history)
UFO_Close_dates['Date']=UFO_Close_dates.index
UFO_Close_dates.info()
```

```python
sns.set()
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
formatter = ticker.FormatStrFormatter('$%1.2f')
ax.yaxis.set_major_formatter(formatter)

plt.xticks(rotation=90)
plt.plot(UFO_history_close)
```

```python
Fidelity = yf.Ticker('FNILX')
Virgin = yf.Ticker('SPCE')
```

```python
Fidelity_history=Fidelity.history(start='2019-04-15')
Fidelity_history_close = pd.DataFrame(Fidelity_history['Close'])

Virgin_history=Virgin.history(start='2019-04-15')
Virgin_history_close = pd.DataFrame(Virgin_history['Close'])


sns.set()
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
formatter = ticker.FormatStrFormatter('$%1.2f')
ax.yaxis.set_major_formatter(formatter)

plt.xticks(rotation=90)
plt.plot(Fidelity_history_close)
plt.plot(Virgin_history_close)
plt.plot(UFO_history_close)
plt.legend(['FNILX','SPCE','UFO'])
plt.title('Comparison of UFO ETF to Market')
```
